```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/fra_cleaned.csv",
    encoding="latin-1",
    sep=";"
)
df.head()


```

```python
rows, cols = df.shape
rows

```

```python
nan_counts = df.isna().sum()
nan_counts
```

```python
unknown_counts = (
    df.select_dtypes(include="object")
      .apply(lambda col: col.str.lower().eq("unknown").sum())
)

unknown_counts
```

```python
TOP_N = 10

for col in df.columns:
    print(f"\n=== {col} ===")
    print(
        df[col]
        .value_counts(dropna=False)
        .head(TOP_N)
    )
```
